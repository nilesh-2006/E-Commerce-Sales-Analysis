```python
import pandas as pd
import numpy as np
```

# Load Clean Dataset

```python
df = pd.read_csv(
    "../data/processed/cleaned_superstore.csv",
    parse_dates=["Order Date","Ship Date"], encoding="latin1"
)
df.head()
```

"parse_dates" isliye use kiya kyunki CSV me date dobara text ban jati hai. Isse Pandas load karte hi unhe datetime me convert kar deta hai

```python
df.shape
df.info()
```

# Create Year Column

```python
df["Year"] = df["Order Date"].dt.year
df.head()
```

# Create Month Column

```python
df["Month"] = df["Order Date"].dt.month
df.head()
```

# Create Month Name

```python
df["Month Name"] = df["Order Date"].dt.month_name()
df.head()
```

# Create Quarter

```python
df["Quarter"] = df["Order Date"].dt.quarter
df
```

# Create Day Name

```python
df["Day Name"] = df["Order Date"].dt.day_name()
df.head()
```

# Create Shipping Days

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
df.head()
```

# Create Profit Margin

```python
df["Profit Margin"] = (df["Profit"] / df["Sales"]) * 100
df
```

# Save Dataset

```python
df.to_csv(
    "../data/processed/featured_superstore.csv",
    index=False
)
df.head()
```
