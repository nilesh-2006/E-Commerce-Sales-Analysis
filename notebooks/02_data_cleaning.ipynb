```python
# Check Outliers (Sales)
Q1 = df["Sales"].quantile(0.25)

Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR

outliers = df[(df["Sales"] < Lower) | (df["Sales"] > Upper)]

print("Total Sales Outliers :", len(outliers))
```

```python
# Check Outliers (Profit)
Q1 = df["Profit"].quantile(0.25)

Q3 = df["Profit"].quantile(0.75)

IQR = Q3 - Q1

Lower = Q1 - 1.5 * IQR

Upper = Q3 + 1.5 * IQR

outliers = df[(df["Profit"] < Lower) | (df["Profit"] > Upper)]

print("Total Profit Outliers :", len(outliers))
```
